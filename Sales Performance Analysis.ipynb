```python
import pandas as pd
import matplotlib .pyplot as plt
```

```python
df =pd.read_csv("ecommerce_sales_data.csv")
```

# Sales Analysis

```python
# Total sales 
```

```python
df['Sales'].sum()
```

```python
# Total profit 
```

```python
df['Profit'].sum()
```

# Product Anaysis

```python
# Top 10 product by sales 
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
```

```python
# Top 10 product by profit 
```

```python
df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False)
```

```python
# lowest performance product 
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=True)
```

```python
# Top categories product 
```

```python
category_sale =df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

# Region Analysis

```python
# Best Region
```

```python
region_sales =df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
# Worst Region
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=True)
```

```python
# Remove Duplicate Values
```

```python
df.drop_duplicates(inplace =True)
```

```python
df['Order Date ']=pd.to_datetime(df['Order Date'])
```

```python
# Create Month column 
```

```python
df['Month']=df['Order Date'].dt.month_name()
```

# Visualization 

```python
# Monthly sales Trend 
```

```python
Monthly_sale =df.groupby('Month')['Sales'].sum().sort_values(ascending=False)
```

```python
Monthly_sale
```

```python
Monthly_sale.plot(kind='line', marker='o', color='purple', figsize=(10, 5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.grid(True)

plt.show()
```

```python
# Sales by Category
```

```python
category_sale.plot(kind='bar')
plt.xlabel("Category")
```

```python
# sales by region 
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(7, 7), cmap='Set2')
plt.title("Sales Distribution by Region", fontsize=14, fontweight='bold')
plt.ylabel("")
plt.show()
```
